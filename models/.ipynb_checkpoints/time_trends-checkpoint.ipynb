```python
%matplotlib inline
from __future__ import division

import pandas as pd
import matplotlib.pyplot as plt
import utils
import datetime

visitors, events, devices, url_categories = utils.load_data(event_categories=True)
```

```python
phirms = events[~events.pharma_firm.isnull()][['dg_id', 'session_id', 'timestamp', 'disease_category', 'disease', 'pharma_firm']]
phirms = phirms.drop_duplicates(subset=['session_id'])
```

```python
phirms['dt'] = phirms.timestamp.apply(lambda x: x.date())
```

```python
fig, ax = plt.subplots(1,1)

phirms.groupby(['dt', 'pharma_firm']).session_id.nunique().reset_index().\
    groupby('pharma_firm').plot(x='dt', y='session_id', ax=ax)
```

```python
# q1=feb/march/april, q2=may/june/july
def get_quarter(dt):
    if dt < datetime.date(2017, 5, 1):
        return 'q1'
    elif dt < datetime.date(2017, 8, 1):
        return 'q2'
```

```python
phirms['quarter'] = phirms.dt.apply(lambda x: get_quarter(x))
```

```python
phirms.groupby('quarter').count()
```

```python
phirmsq = phirms.groupby(['pharma_firm', 'quarter']).session_id.nunique().reset_index()
phirmsq = phirmsq.pivot(index='pharma_firm', columns='quarter', values='session_id').reset_index().fillna(0)
phirmsq.columns = ['pharma_firm', 'q1_session_count', 'q2_session_count']
```

```python
phirmsq.head()
```

```python
phirmsdt = phirms.groupby(['pharma_firm', 'dt']).session_id.nunique().reset_index()
phirmsdt.columns = ['pharma_firm', 'date', 'session_count']
```

```python
phirmsdt.head()
```

```python
phirmsq.to_csv('../data/pharma_firms_by_quarter.csv', index=False)
phirmsdt.to_csv('../data/pharma_firms_by_date.csv', index=False)
```
